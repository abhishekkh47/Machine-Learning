```python
import pandas as pd
```

```python
dataset = pd.read_excel('dataset.xlsx')
```

```python
place = dataset[['X','Y']]
```

```python
place=place.values.tolist()
```

```python
map_india = folium.Map(location=[20.0102299,64.4287732] ,control_scale=True, zoom_control=True, zoom_start=5 )
```

```python
i=0
for points in place:
    folium.Marker(location=points ,
                  popup= dataset['Hospital_Name'][i],
                  icon= folium.Icon(color='red', icon = 'tint',icon_color='white')).add_to(map_india)
    i+=1
```
