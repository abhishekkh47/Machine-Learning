# K-Nearest Neighbour ( KNN ) Algorithm:
- When we have a test data, then we check the nearest neighbours.
- We need to specify the number of neighbours (k) [k-value] , we need to check.
- We always give the "k-value" as odd-integers.

### KNN may be used for Classification and Regression. But , usually we use it for Classification
### The only challenge in KNN is that, it is very slow and very compute intensive.
### Therefore, it is not at all recommended for Big Data

## Why to use KNN ?
- We need not to train the model.
- Linear and Logictic Regression Algos ( Regression and Classification ) is generally used for Linear Data.
  But Real World problem are mostly Non-Linear. And KNN is better for that (Small data).
- We should use Neural Networks for BigData.

```python
import pandas as pd
```

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
```

```python
dataset.info()
```

```python
dataset.columns
```

```python
X = dataset[['Age', 'EstimatedSalary']]
y = dataset['Purchased']
```

```python
import seaborn as sns
```

```python
sns.set()
```

```python
sns.scatterplot(x='Age', y='EstimatedSalary', data=dataset)
```

```python
type(X)
```

```python
# X is in Pandas Dataframe, we need to convert into numpy
X = X.values
y = y.values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=25)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Create the model
model = KNeighborsClassifier(n_neighbors=15)
# we need to change the value of neighbours to get best results
# Here it seems like we are training the model but actually we are not.
# We are just inserting the data into it.
```

```python
model.fit(X_train, y_train)
```

```python
model.predict([[50, 200000]])
```

```python
model.predict([[20, 20000]])
```

```python
# To predict all test cases
y_pred = model.predict(X_test)
```

```python
# to find accuracy, we use confusion matrix
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test ,y_pred)
```

```python
# we can create a loop to change the k-value and check the accuracy
# Store the accuracy in a list
```

```python
import numpy as np
```

```python
error_rate = []
for i in range(1,50):
    model = KNeighborsClassifier(n_neighbors=i)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    error_rate.append(np.mean(y_test != y_pred))
    #error_rate.append(accuaccuracy_score(y_test, y_pred))
```
