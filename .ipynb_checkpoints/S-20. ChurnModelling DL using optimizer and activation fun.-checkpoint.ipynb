```python
import pandas as pd
```

```python
dataset = pd.read_csv('Churn_Modelling.csv')
```

```python
dataset.columns
```

```python
# X = dataset[['CreditScore','Geography','Gender','Age','Tenure','Balance','NumOfProducts','HasCrCard','IsActiveMember','EstimatedSalary']]
X = dataset[['CreditScore','Age','Balance', 'EstimatedSalary']]
```

```python
y = dataset['Exited']
```

```python
# we not need to do feature selection/Elimination
# but we need to do feature engineering
```

```python
geo = dataset['Geography']
geo = pd.get_dummies(geo, drop_first=True)
```

```python
gender = dataset['Gender']
gender = pd.get_dummies(gender, drop_first=True)
```

```python
numofproducts = dataset['NumOfProducts']
numofproducts = pd.get_dummies(numofproducts, drop_first=True)
```

```python
crcard = dataset['HasCrCard']
crcard = pd.get_dummies(crcard, drop_first=True)
```

```python
active = dataset['IsActiveMember']
active = pd.get_dummies(active, drop_first=True)
```

```python
estdsal = dataset['EstimatedSalary']
estdsal = pd.get_dummies(estdsal, drop_first=True)
```

```python
tenure = dataset['Tenure']
tenure = pd.get_dummies(tenure, drop_first=True)
```

```python
X
```

```python
tenure
```

```python
X = pd.concat([X, geo, gender, numofproducts, crcard, active, tenure], axis=1)
```

```python
X.isnull()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
y_train
```

```python
from keras.models import Sequential
```

### Here, we are using "Sequential" as our output is classification based and we are sequencing the neurons

```python
import seaborn as sns
```

```python
age = X['Age']
```

```python
sns.scatterplot(x=age, y=y,)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(age,y)
```

```python
# create model
model = Sequential()
```

```python
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
model.add(Dense(units=6, input_dim=22, activation='relu'))
# used 'relu' for cleaning the data, avoid random increase/decrease in loss
```

```python
model.add(Dense(units=6, activation='relu'))
```

```python
model.add(Dense(units=6, activation='relu'))
```

```python
# create output layer of 1 unit for output neuron
model.add(Dense(units=1, activation='sigmoid'))
```

```python
# Now we compile the code : create one output architecture
# MSE => Regression
# Binary_crossentropy => Binary classification
# Categorical_crossetropy => Multi classification
model.compile(optimizer=Adam(learning_rate=0.000001), loss='binary_crossentropy')
```

```python
# now train the model
model.fit(X_train,y_train, epochs=50)
# increase number of epochs or Dense units to increase the accuracy
```

```python
model.summary()
```

```python
df_loss = pd.DataFrame(model.history.history)
```
